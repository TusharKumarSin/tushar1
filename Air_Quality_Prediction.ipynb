```python
air_quality_data=air_quality_data.replace(to_replace=-200,value=np.NaN)
```

```python
numerical = air_quality_data.select_dtypes(include=['number'])
mean_values = numerical.mean()
```

```python
air_quality_data=air_quality_data.fillna(mean_values)
```

FB Prophet docunentation

```python
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info=air_quality_data['Time']
```

```python
print(time_info)
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
print(date_time.head())
```

```python
date_time.shape
```

```python
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['CO(GT)']
```

```python
data=data.drop(['CO'],axis=1)
```

```python
data.head()
```

```python
from prophet import Prophet
```

 Traning Model

```python
model=Prophet()
```

```python
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq='D')
```

```python
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
